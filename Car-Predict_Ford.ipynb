```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv("ford.csv")
data.head()
```

```python
df = data[["year","price","mileage","tax","mpg","engineSize"]]
df
df = df.sort_values("price",ascending=False)
df.head(20)
df = df[10:]
df
```

```python
x = df[["year","mileage","tax","mpg","engineSize"]].values
y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(64, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(128, activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x_train, y_train,epochs=300,batch_size=32)
```

```python
df2 = df
df3 = df
```

```python
yeniaraba = df.drop("price",axis=1)
```

```python
yeniaraba = scaler.transform(yeniaraba.values.reshape(-1,5))
```

```python
sonuc = model.predict(yeniaraba)
sonuc
```
